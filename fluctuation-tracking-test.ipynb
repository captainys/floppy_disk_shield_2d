```python
import math

data = []
angle = 0
cell = 8
for i in range(1000):
    data.append(16 + math.sin(angle)*5)
    angle += 0.01

for dt in data:
    integer = int(dt / cell + 0.5)
    error = dt - integer*cell
    cell += error * 0.1
    print('{:8.3} {:8.3} {:4} {:8.3}'.format(dt, cell, integer, error))
```
